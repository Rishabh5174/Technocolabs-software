# Feature Engineering

```python
df.info()
```

```python
df['Dept_flights_time'].value_counts()  
```

```python
# Convert datatype 'object'  to 'integer' for features 'Dept_flights_time'.

df['Dept_flights_time']=df['Dept_flights_time'].map({'Morning_flight':1,'Afternoon_flight':2,'Evening_flight':3,'Night_flight':4}).astype('int')
```

```python
df['Dept_flights_time']
```

```python
df['Cabin'].unique()
```

```python
# Convert "Cabin" datatype from object to integer by using mapping.

df['Cabin']=df['Cabin'].map({'E':1,'B':2,'PE':3}).astype('int')
```

```python
df['Cabin']
```

```python
df.info()
```

### Encoding Categorical Variables
#### Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
var_mode = df.select_dtypes(include ='object').columns

# For  Dataset
for i in var_mode:
    df[i] = le.fit_transform(df[i])
```

```python
# Showing the dataset after performing Label Encoding

df.head(10)
```

```python
df.info()
```

```python
# correlation matrix

plt.figure(figsize=(12,10))
sns.heatmap(df.corr(),annot=True)
```

```python
# Divide dataset into dependent and independent features

# independent features
x=df.drop('optimal_hours',axis=1)

# dependent feature
y=df['optimal_hours']
```

```python
y.dtypes
```

### Feature scailing 

```python
# Apply feature scailing to the independent features:

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
x_scaled
```

```python
# splitting the data into train set and test set 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=101) # 80:20 ratio
```

```python
# Applying K-Nearest Neighbor Regressor.
from sklearn.neighbors import KNeighborsRegressor
#Knn Model Creation
knn=KNeighborsRegressor(n_neighbors=5,metric='minkowski',p=2)
model_knn=knn.fit(x_train,y_train)
```

```python
# prediction
y_pred=model_knn.predict(x_test)
```

```python
# Checking R-Squared value
from sklearn.metrics import r2_score
r2=r2_score(y_test,y_pred)
print('R-squared value of the model: ',r2)
```

```python
# Checking Root Mean Squared Error
from sklearn.metrics import mean_squared_error
MSE =mean_squared_error(y_test, y_pred)
rmse = np.sqrt(MSE)
print('RSME value of the model: ',rmse)
```

# Apply PCA (Principle Component Analysis)

```python
from sklearn.decomposition import PCA
pca=PCA(0.95) # components store 95% of the information
x_pca=pca.fit_transform(x_scaled)
x_pca.shape
```

```python
# Showing the components storing how much variations

pca.explained_variance_ratio_
```

```python
# no of principle components

pca.n_components_
```

```python
#Now we know the components are 10 but we can plot 2 or 3d to see their distribution throught visualization 
plt.figure(figsize=(10,10))
plt.scatter(x_pca[:,0],x_pca[:,1],c=y,cmap='plasma')
plt.xlabel('pc1')
plt.ylabel('pc2')
```

#### Splitting the data in train set and test set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_pca,y,test_size=0.33,random_state=101)
```

# Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
```

#### Baseline algorithm performance

```python
rf=rf.fit(x_train,y_train)
```

```python
y_pred=rf.predict(x_test)
```

#### Checking explained variance score

```python
from sklearn.metrics import explained_variance_score
score=explained_variance_score(y_test,y_pred)
print('Model explianed variance score: ',score)
```

#### Checking R-Squared value

```python
from sklearn.metrics import r2_score
r2=r2_score(y_test,y_pred)
print('R-squared value of the model: ',r2)
```

#### Checking Root Mean Squared error(RMSE) of the model

```python
from sklearn.metrics import mean_squared_error
MSE =mean_squared_error(y_test, y_pred)
rmse = np.sqrt(MSE)
print('RSME value of the model: ',rmse)
```

```python
rf.feature_importances_
```

```python
sorted_idx = rf.feature_importances_.argsort()
plt.barh(df.columns[sorted_idx], rf.feature_importances_[sorted_idx])
plt.xlabel("Random Forest Feature Importance")
```

So we don't need to fine-tune 'max_depth' and 'n_estimators' parameters for our RandomForest Regressor by using GridSearch.

# Lasso Regression

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_pca,y,test_size=0.33,random_state=101)
```

```python
from sklearn.linear_model import Lasso
lasso=Lasso(alpha=0.0001,normalize=True)
lasso.fit(x_train,y_train)
```

```python
y_pred=lasso.predict(x_test)
y_pred
```

```python
# slope

print('Slope: %.2f'%lasso.coef_[0])
```

```python
# intercept

print('Intercept: %.2f'%lasso.intercept_)
```

```python
# performance of the model

# MSE
from sklearn.metrics import mean_squared_error

print('Mean Squared Error: ',mean_squared_error(y_test,y_pred))
```

```python
# RMSE
print('RSME value of the model: ',np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
# R-squared
from sklearn.metrics import r2_score
r2=r2_score(y_test,y_pred)
print('R-squared value of the model: ',r2)
```
